```python
# import vulnerability data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

vulnerability_data = pd.read_csv('vulnerabilityroadN1.csv')
vulnerability_data.head()
economic_vulnerability = vulnerability_data['TotalEconomicVulnerability'].tolist()
road_segments = vulnerability_data['Unnamed: 0'].tolist()

part_roadsegments = road_segments[0:100]
part_vulnerability = economic_vulnerability[0:100]
```

```python
# without segment length, horizontal

number_segments = len(road_segments)
print(number_segments)

x = np.arange(number_segments)

y = np.ones(number_segments)

xmin = x.min()
xmax = x.max()

fig, axs = plt.subplots(ncols=1, sharey=True, figsize=(20, 1.5))
fig.subplots_adjust(hspace=0.5, left=0.07, right=0.93)
hb = axs.hexbin(x, y, economic_vulnerability, gridsize=number_segments-1, cmap='RdYlGn_r')
axs.axis([xmin, xmax, 0.9999, 1.0001])
axs.get_yaxis().set_visible(False)
plt.xticks(x,road_segments)
plt.xlabel('Road Segment')
cb = fig.colorbar(hb, ax=axs, orientation='horizontal', pad=0.5)
cb.set_label('Vulnerability')

plt.show()
```

```python
# without segment length, horizontal, only part of the data

number_segments = len(part_roadsegments)
print(number_segments)

x = np.arange(number_segments)

y = np.ones(number_segments)

xmin = x.min()
xmax = x.max()

fig, axs = plt.subplots(ncols=1, sharey=True, figsize=(20, 1.5))
fig.subplots_adjust(hspace=0.5, left=0.07, right=0.93)
hb = axs.hexbin(x, y, part_vulnerability, gridsize=number_segments-1, cmap='RdYlGn_r')
axs.axis([xmin, xmax, 0.9999,1.0001])
axs.get_yaxis().set_visible(False)
plt.xticks(x,road_segments)
plt.xlabel('Road Segment')
cb = fig.colorbar(hb, ax=axs, orientation='horizontal', pad=0.5)
cb.set_label('Vulnerability')

plt.show()
```
